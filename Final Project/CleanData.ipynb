```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Docked_Bikeshare_Ridership_by_System__Year__and_Month.csv")

# Drop columns
df = df.drop(columns=["the_geom", "sysorder", "mar_ind", "may_ind", 
                      "dateStr", "dateTxt", "bin_num_trip", 
                      "yroverchange_bin", "display", "sysname_alt", "trip_type"])

# Save the cleaned dataset
df.to_csv("DockedBikeRidersdip_Clean.csv", index=False)
```

```python
import pandas as pd

# Load the CSV into a DataFrame
df = pd.read_csv("DockedBikeRidersdip_Clean.csv")

# Update `pct_ttl_num_trips` to 0 where `sum_min` and `num_trip` are both 0
df.loc[(df['sum_min'] == 0) & (df['num_trip'] == 0), 'pct_ttl_num_trips'] = 0

# Save the updated DataFrame back to a CSV
df.to_csv("DockedBikeRidersdip_Updated.csv", index=False)
```

```python
import pandas as pd

# Load the CSV into a DataFrame
df = pd.read_csv("DockedBikeRidersdip_Clean.csv")

# Update `pct_ttl_num_trips` and `yroverchange` to 0 where `sum_min` and `num_trip` are both 0
df.loc[(df['sum_min'] == 0) & (df['num_trip'] == 0), ['pct_ttl_num_trips', 'yroverchange']] = 0

# Save the updated DataFrame back to a CSV
df.to_csv("DockedBikeRidersdip_Updated.csv", index=False)
```

```python
import pandas as pd

# Load the original CSV file
df = pd.read_csv('DockedBikeRidersdip_Updated.csv')
# Extract the relevant columns for the system_table
system_table_df = df[['sysid', 'sysname', 'lat', 'lng']].drop_duplicates()

# Save to a new CSV file
system_table_df.to_csv('system_table.csv', index=False)
```

```python
# Extract the relevant columns for the date_table
date_table_df = df[['year', 'assigned_month']].drop_duplicates()

# Save to a new CSV file
date_table_df.to_csv('date_table.csv', index=False)
```

```python
# Extract the relevant columns for the fact_ridership table
fact_ridership_df = df[['sysid', 'year', 'assigned_month', 'sum_min', 'num_trip', 'pct_ttl_num_trips', 'yroverchange']]

# Save to a new CSV file
fact_ridership_df.to_csv('fact_ridership.csv', index=False)
```
